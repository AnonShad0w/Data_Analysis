```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Satellite Database.csv')
```

```python
dfssl=df.set_index("Contractor")
```

```python
dfssl.loc['Space Systems/Loral']
```

```python
dfssl['Year'] = pd.DatetimeIndex(dfssl['Date of Launch']).year
```

```python
dfssl.loc['Space Systems/Loral']
```

```python
dfssl.head()
```

```python
plt.figure(figsize=(10,15))
ax = sns.catplot(y='Year', data=dfssl.loc['Space Systems/Loral'], kind='count')

ax.set_xticklabels(rotation=40, ha="right")
plt.tight_layout()
```

```python
plt.figure(figsize=(10,15))
ax = sns.lmplot(x='', y='Year', data=dfssl.loc['Space Systems/Loral'])

ax.set_xticklabels(rotation=40, ha="right")
plt.tight_layout()
```
